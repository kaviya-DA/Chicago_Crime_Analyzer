```python
#  Convert 'Date' and 'Updated On' to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Updated On'] = pd.to_datetime(df['Updated On'], errors='coerce')

#  Extract useful time columns from 'Date'
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Hour'] = df['Date'].dt.hour
df['DayOfWeek'] = df['Date'].dt.day_name()

#  Convert 'Arrest' and 'Domestic' to boolean (True/False)
df['Arrest'] = df['Arrest'].astype(bool)
df['Domestic'] = df['Domestic'].astype(bool)

#  Drop rows where Latitude or Longitude is missing
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)

#  Fill missing values in location-related columns (optional)
df['Community Area'].fillna(df['Community Area'].mode()[0], inplace=True)
df['Ward'].fillna(df['Ward'].mode()[0], inplace=True)
df['District'].fillna(df['District'].mode()[0], inplace=True)
df['Beat'].fillna(df['Beat'].mode()[0], inplace=True)

#  Final check
print(" Cleaned Data Shape:", df.shape)
print(" Missing values after cleaning", df.isnull().sum())
```

```python
import numpy as np

 # Step 1: Replace 'NA' with None so SQL understands
df = df.replace('NA', np.nan)
df = df.replace({np.nan: None})

# Step 2: Keep only 22 original columns (in correct order)
required_columns = [
    'ID', 'Case Number', 'Date', 'Block', 'IUCR', 'Primary Type', 'Description', 'Location Description',
    'Arrest', 'Domestic', 'Beat', 'District', 'Ward', 'Community Area', 'FBI Code',
    'X Coordinate', 'Y Coordinate', 'Year', 'Updated On', 'Latitude', 'Longitude', 'Location'
]

df = df[required_columns]

#  Step 3: Convert to list of tuples
data_to_insert = df.values.tolist()

#  Step 4: Insert query with 22 columns and 22 %s placeholders
insert_query = """
INSERT INTO crime_data (
    ID, Case_Number, Date, Block, IUCR, Primary_Type, Description, Location_Description,
    Arrest, Domestic, Beat, District, Ward, Community_Area, FBI_Code,
    X_Coordinate, Y_Coordinate, Year, Updated_On, Latitude, Longitude,
    Location
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

#  Step 5: Insert data in batches
batch_size = 1000

for i in range(0, len(data_to_insert), batch_size):
    batch = data_to_insert[i:i + batch_size]
    try:
        cursor.executemany(insert_query, batch)
        mydb.commit()
        print(f"Successfully inserted batch {i} to {i + len(batch)}")
    except Exception as e:
        print(f" Error in batch {i} to {i + len(batch)}: {e}")
        mydb.rollback()
        break
```
